```python
import numpy as np
from matplotlib import pyplot as plt 
from scipy import ndimage
%matplotlib inline

datafile = open(r'D:\GitHub\Neural-Network\mnist_dateset\mnist_test.csv', 'r')
datalist = datafile.readlines()
datalist = datalist[1:]
datefile.close()

record = 1

allvalues = datelist[record].split(',')
print('lable is', allvalues[0])

# scale the input 
input_array = ((np.asfarray(allvalues[1:]) * 0.99 / 255) + 0.01).reshape((28,28))

# plot the image
plt.imshow(input_array, cmap = 'Greys', interpolation= None)

# create rotated variations 
inputs_plus10_img = ndimage.rotate(input_array, 10.0, cval = 0.01, order = 1, reshape= False)
inputs_minus10_img = ndimage.rotate(input_array, - 10.0, cval = 0.01, order =1, reshape = False)
```

```python
# plot the rotated images 
plt.imshow(inputs_plus10_img, interpolation= None)
```
