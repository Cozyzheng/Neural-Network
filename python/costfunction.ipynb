```python
import numpy as np 
from matplotlib import pyplot as plt 

def sigmoid(x):
    y = 1 / (1 + np.exp(x))
    return y 

def tanh(x):
    y = (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))
    return y 
x = np.linspace(-100,100, 10000)

def ReLU(x):
    return np.maximum(0,x)

def cross_entropy_error(y, p):
    delta = 1e-7
    return np.sum(-y * np.log(p + delta)) 

y = ReLU(x)
plt.plot(x, y)

y = np.array([0, 1, 0, 0])
p = np.array([0.3, 0.2, 0.1, 0.4])
p1 = np.array([0, 1, 0, 0])
print(cross_entropy_error(y, p1))
```
